```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import newton
```

```python
def T(x, y):
    n = len(x)-1
    v = 0
    for i in range(n):
        v += np.sqrt(((x[i+1]-x[i])**2 + (y[i+1]-y[i])**2)/y[i])
    v *= 0.22587697572
    return v

def draw(x, y, label):
    plt.axis('equal')
    plt.plot(x, -y, '-', label=label)
    plt.legend(framealpha=0.25, frameon=False)
    print(T(x,y))

def cicloide(a,b):
    # Trobem l'angle final numèricament, despres R
    def angle(theta):
        return b/a - (1-np.cos(theta))/(theta-np.sin(theta))
    theta_final = newton(angle, np.pi/2)
    print("theta_final=",theta_final)
    
    R = b/(1-np.cos(theta_final))
    Fx = lambda t: R*(t - np.sin(t))
    Fy = lambda t: R*(1 - np.cos(t))
    t = np.linspace(0, theta_final, 50)
    x = Fx(t)
    y = Fy(t)
    x = np.delete(x, [0])
    y = np.delete(y, [0])
    print(x[-1], np.sqrt(a**2+b**2), R*(theta_final - np.sin(theta_final)))
    draw(x, y, label="cicloide")
    
```
